## MIL object can be used to perform segmentation
### Three functions to note:
* setImage(path)
* performLabelling(listVocSegmented)
* computeCategoryAcc(category, #image, dataset)

setImage() defines the next image that will be labelled using performLabelling.
ComputeCategoryAcc() computes the validation score of a category.

## Run MIL on chosen images with VOC segmentations:

Call performLabelling() with a list of name of VOC image

You can call retCatFileNames("catname") which will give you a list of all images of a certain category from the valdataset

```python
# MIL() will load all val images so takes a few second
mil = MIL()
mil.setImage("data/VOC/VOCdevkit/VOC2012/JPEGImages/2011_002447.jpg")
crf, normal = mil.performLabelling(mil.retCatFileNames("sheep"))
```

### Show result

```python
im = load_image2("data/VOC/VOCdevkit/VOC2012/JPEGImages/2011_002447.jpg")
show_masks(im, im, normal.transpose((2,0,1)), np.array(crf)[:,0,...], K=21)
```
