```python

class LoggedSequential(nn.Module):
    def __init__(self, modules_dict, log_layers=None):
        super(LoggedSequential, self).__init__()
        self.log_layers = log_layers
        for key, module in modules_dict.items():
            self.add_module(key, module)
        self.output = None

    def forward(self, x):
        self.output = OrderedDict()
        for name, module in self._modules.items():
            x = module(x)
            if self.log_layers is None or name in self.log_layers:
                self.output[name] = x
        return x

    def flush(self):
        self.output = None

class LoggedNet(nn.Module):
    def __init__(self, net, log_features=None):
        super(LoggedNet, self).__init__()
        self.features = LoggedSequential(net.features._modules, log_layers=log_features)
        self.classifier = None
        if net.classifier is not None:
            self.classifier = LoggedSequential(net.classifier._modules)

    def forward(self, x):
        x = self.features.forward(x)
        if self.classifier is not None:
            x = x.view(x.size(0), -1)
            x = self.classifier.forward(x)

        return x

    def flush(self):
        self.features.flush()
        if self.classifier is not None:
            self.classifier.flush()
            
vgg19_bn_feature_index = {
    'conv1_1': '2',
    'conv1_2': '5',
    'conv2_1': '9',
    'conv2_2': '12',
    'conv3_1': '16',
    'conv3_2': '19',
    'conv3_3': '22',
    'conv3_4': '25',
    'conv4_1': '29',
    'conv4_2': '32',
    'conv4_3': '35',
    'conv4_4': '38',
    'conv5_1': '42',
    'conv5_2': '45',
    'conv5_3': '48',
    'conv5_4': '51'
}

vgg16_feature_index = OrderedDict((('conv1_2', '3'),
                                    ('conv2_2', '8'),
                                    ('conv3_3', '15'),
                                    ('conv4_3', '22'),
                                    ('conv5_3', '29')))

vgg16_bn_feature_index = OrderedDict((('conv1_2', '5'),
                                     ('conv2_2', '12'),
                                     ('conv3_3', '22'),
                                     ('conv4_3', '32'),
                                     ('conv5_3', '42')))

class HypercolumnExtractor:
    def __init__(self, net, layer_names):
        self.net = LoggedNet(net, log_features=layer_names)

    def get(self, x):
        from torch.nn.functional import upsample
        self.net.features.forward(x)
        input_size = x.size()[2:]
        output = []
        for k, v in self.net.features.output.items():
            #output.append(upsample(self.net.features.output[k], input_size, mode='bilinear'))
            output.append(self.net.features.output[k])
        largest_size = output[np.array([f.size(2) for f in output]).argmax()].size()[2:]
        for i in range(len(output)):
            output[i] = upsample(output[i], largest_size, mode='bilinear')
        
        hypercols = torch.cat(output, dim=1)
        return hypercols

    def flush(self):
        self.net.flush()
```

```python
def get_original_vgg16(path_to_params='/home/ecollins/original_vgg16_pytorch/vgg16-00b39a1b.pth'):
    from torchvision.models.vgg import vgg16
    original_vgg16 = vgg16()
    original_params = torch.load(path_to_params)
    original_params['classifier.0.weight'] = original_params['classifier.1.weight']
    original_params['classifier.0.bias'] = original_params['classifier.1.bias']
    original_params['classifier.3.weight'] = original_params['classifier.4.weight']
    original_params['classifier.3.bias'] = original_params['classifier.4.bias']
    del original_params['classifier.1.weight']
    del original_params['classifier.1.bias']
    del original_params['classifier.4.weight']
    del original_params['classifier.4.bias']
    original_vgg16.load_state_dict(original_params)
    return original_vgg16
```

```python
def preprocess(img):
    img = img.astype('float32')/255
    mean=np.array([0.485, 0.456, 0.406])
    std=np.array([0.229, 0.224, 0.225])
    img -= mean[None, None, :]
    img /= std[None, None, :]
    img = img.transpose(2,0,1)
    return img

#def preprocess(img):
#    img = img.astype('float32')
#    original_vgg16_mean = np.array([103.939, 116.779, 123.68]).astype('float32') #given in BGR
#    img = img[:, :, ::-1] - original_vgg16_mean
#    img = img.transpose(2, 0, 1)
#    return img

def get_img_features(img, layers=None):
    vgg.features(Variable(torch.from_numpy(img[None])))
    if layers is None:
        features = {k: vgg.features.output[v][0].data.numpy() for k,v in vgg16_feature_index.items()}
    else:
        features = {l: vgg.features.output[vgg16_feature_index[l]][0].data.numpy() for l in layers}
    return features

def get_img_features(img):
    img = Variable(torch.from_numpy(img[None]))
    return hypercolumns.get(img)[0].data.numpy()
```

```python
def get_association_matrix(features1, features2):
    '''
    :param features1: CxHxW
    :param features2: CxH'xW'
    :return: (H*W)x(H'*W')
    '''
    assert features1.shape[0] == features2.shape[0]
    C = features1.shape[0]
    features1 = features1.reshape(C,-1)
    features2 = features2.reshape(C,-1)
    return np.dot(features1.transpose(), features2)


def get_nmf_maps(features1, features2, K=5, offset=0, l1_ratio=0, alpha=0, random_state=0, max_iter=200):
    '''
    :param features1: CxHxW
    :param features2: CxH'xW'
    :param K: Number of maps
    :return: KxHwW, KxH'xW'
    '''
    association_matrix = get_association_matrix(features1, features2)
    
    from sklearn.decomposition import NMF
    model = NMF(n_components=K, init='random', random_state=random_state, l1_ratio=l1_ratio, alpha=alpha, max_iter=max_iter)
    W = model.fit_transform(association_matrix)
    W = W.transpose().reshape((K, features1.shape[1], features1.shape[2]))
    
    H = model.components_
    H = H.reshape((K, features2.shape[1], features2.shape[2]))
    
    return W, H
```
